```python
# ===============================
# üìä PROJET : ANALYSE BANK MARKETING (UCI)
# Auteur : Douae El Fagrouch
# Date : Novembre 2025
# ===============================

# ===============================
# 1Ô∏è‚É£ Pr√©sentation du sujet
# ===============================

from IPython.display import Markdown, display

display(Markdown("""
# Analyse du jeu Bank Marketing (UCI)

## Contexte
Ce projet porte sur l'analyse du jeu de donn√©es Bank Marketing provenant de l'UCI.
Ce dataset contient des informations sur des campagnes marketing t√©l√©phoniques
d'une banque portugaise et permet de pr√©dire si un client va souscrire un d√©p√¥t √† terme.

## Objectif
- Comprendre les tendances g√©n√©rales du dataset
- Identifier les corr√©lations entre les variables
- D√©terminer le profil des clients susceptibles de souscrire
- Visualiser les distributions et comportements des variables

## Contenu du projet
- Analyse statistique : moyenne, m√©diane, variance, √©cart-type, min/max
- Visualisations : histogrammes, scatterplots, boxplots, barplots, heatmap
- Interpr√©tation des r√©sultats et conclusions
"""))

# ===============================
# 2Ô∏è‚É£ Import des biblioth√®ques
# ===============================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from ucimlrepo import fetch_ucirepo

# ===============================
# 3Ô∏è‚É£ Chargement du jeu de donn√©es
# ===============================
bank_marketing = fetch_ucirepo(id=222)

# Variables explicatives et cible
X = bank_marketing.data.features
y = bank_marketing.data.targets

# Fusion
df = pd.concat([X, y], axis=1)

print("‚úÖ Jeu de donn√©es charg√© avec succ√®s")
print(df.head())

# ===============================
# 4Ô∏è‚É£ Informations g√©n√©rales
# ===============================
print("\n--- Informations sur le dataset ---")
print(df.info())
print("\nDimensions :", df.shape)
print("\nValeurs manquantes par colonne :")
print(df.isnull().sum())

# ===============================
# 5Ô∏è‚É£ Analyse descriptive des variables num√©riques
# ===============================
numeric_df = df.select_dtypes(include=[np.number])
print("\n--- Statistiques descriptives ---")
print(numeric_df.describe())

print("\n--- Moyenne / M√©diane / Ecart-type / Variance ---")
for col in numeric_df.columns:
    print(f"\nüìà {col} :")
    print(f"  Moyenne : {numeric_df[col].mean():.2f}")
    print(f"  M√©diane : {numeric_df[col].median():.2f}")
    print(f"  Variance : {numeric_df[col].var():.2f}")
    print(f"  Ecart-type : {numeric_df[col].std():.2f}")
    print(f"  Min : {numeric_df[col].min():.2f}")
    print(f"  Max : {numeric_df[col].max():.2f}")
    print(f"  Q1 : {numeric_df[col].quantile(0.25):.2f}")
    print(f"  Q2 (m√©diane) : {numeric_df[col].quantile(0.50):.2f}")
    print(f"  Q3 : {numeric_df[col].quantile(0.75):.2f}")
    print(f"  IQR : {numeric_df[col].quantile(0.75)-numeric_df[col].quantile(0.25):.2f}")

# ===============================
# 6Ô∏è‚É£ Analyse de la variable cible
# ===============================
target = 'y'
print("\n--- R√©partition de la variable cible ---")
print(df[target].value_counts(normalize=True)*100)

plt.figure(figsize=(5,4))
sns.countplot(x=target, data=df, palette="pastel")
plt.title("R√©partition de la variable cible (Souscription au d√©p√¥t √† terme)")
plt.xlabel("Souscription (oui/non)")
plt.ylabel("Nombre d'observations")
plt.show()

# ===============================
# 7Ô∏è‚É£ Visualisations et tendances
# ===============================
# Histogramme √¢ge
plt.figure(figsize=(7,4))
sns.histplot(df['age'], bins=30, kde=True, color='skyblue')
plt.title("Distribution de l'√¢ge des clients")
plt.xlabel("√Çge")
plt.ylabel("Nombre de clients")
plt.show()

# Boxplot solde
plt.figure(figsize=(7,4))
sns.boxplot(x=df['balance'], color='lightgreen')
plt.title("Valeurs extr√™mes du solde moyen des clients")
plt.show()

# Scatterplot √¢ge vs solde
plt.figure(figsize=(7,4))
sns.scatterplot(x='age', y='balance', data=df, alpha=0.6)
plt.title("Relation entre √¢ge et solde moyen")
plt.xlabel("√Çge")
plt.ylabel("Solde moyen")
plt.show()

# Barplot niveau d'√©ducation
plt.figure(figsize=(7,5))
df['education'].value_counts().plot(kind='bar', color='coral')
plt.title("R√©partition du niveau d'√©ducation")
plt.xlabel("Niveau d'√©ducation")
plt.ylabel("Nombre de clients")
plt.show()

# ===============================
# 8Ô∏è‚É£ Corr√©lations
# ===============================
plt.figure(figsize=(10,6))
corr_matrix = numeric_df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Matrice de corr√©lation")
plt.show()

# ===============================
# 9Ô∏è‚É£ Interpr√©tation rapide
# ===============================
print("""
üß† Conclusion rapide :
- La majorit√© des clients ont entre 30 et 50 ans.
- La variable cible (souscription) est d√©s√©quilibr√©e : peu de clients souscrivent.
- Le solde moyen varie fortement, avec quelques valeurs extr√™mes.
- Le niveau d'√©ducation le plus fr√©quent est 'secondary'.
- Les corr√©lations entre variables num√©riques sont faibles √† mod√©r√©es.
""")
```
